```python
## Python Tuples
# Access Tuple Items
Brands = ("BMW", "KTM" , "Honda")
print(Brands[1:3])
```

```python
# This Error Shows that replacement is not allowed
Brands = ("BMW", "KTM" , "Honda")
Brands [2] = "Kawasaki"
print(Brands)
```

```python
# Append Items (Tuples dont allow any modification)
Brands = ("BMW", "KTM" , "Honda")
Brands.append("Kawasaki")
print(Brands)
```

```python
#how to use multiple tuples and how to combine them
t1 = (1,2,3)
t2 = (4,5)
t3 = t2 + t1
print("t1:", t1)
print("t2:", t2)
print("t3:", t3)
```

```python
# Result and grade and distinction allocator using tuple
result = ("Maths", 82 , "A")
subject , marks, grade = result
print("Subject:", subject)
print("Marks:", marks)
print("Grade:", grade)
if marks>=75:
  print("Distinction Obtained")
```

```python
# Monitoring Attendance of Employees using tuple
attendance = ("P", "A", "A", "A" , "P" , "A")
print("Total Present", attendance.count("P"))
print("Total Absent", attendance.count("A"))
if "A" in attendance:
  print("Attendance is not good")
```
